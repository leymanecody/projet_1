## Voici l'Objectif du point de contrôle : Extraire du texte du site Web de Wikipedia à l'aide de Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup

def get_html_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup

soup = get_html_content('https://fr.wikipedia.org/wiki/Python_(langage)')

# extraire le titre 
def get_article_title(soup):
    title = soup.find('h1', {'class': 'firstHeading'}).text
    return title
title = get_article_title(soup)
print(title)

#Pour extraire le texte de l'article pour chaque paragraphe avec leurs rubriques.
#Associez ces titres à leurs paragraphes respectifs dans le dictionnaire

def get_article_text(soup):
    article_text = {}
    paragraphs = soup.find_all('p')
    for p in paragraphs:
        if p.find('b'):
            section_title = p.find('b').text
            section_text = p.text.replace(section_title, '')
            article_text[section_title] = section_text
    return article_text
# exécutons le code 

article_text = get_article_text(soup)
for title, text in article_text.items():
    print(title)
    print(text)


```

```python
# Pour collecter chaque lien qui redirige vers une autre page Wikipédia 
def get_wikipedia_links(soup):
    links = []
    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('/wiki/'):
            links.append(href)
    return links
# exécutons le code
links = get_wikipedia_links(soup)
for link in links:
    print(link)
```

```python
#  Pour envelopper toutes les fonctions précédentes dans une seule fonction qui prend comme paramètres un lien Wikipedia
def get_wikipedia_article(url):
    soup = get_html_content(url)
    title = get_article_title(soup)
    article_text = get_article_text(soup)
    links = get_wikipedia_links(soup)
    return {'title': title, 'text': article_text, 'links': links}
# exécutons le code 
article_data = get_wikipedia_article('https://fr.wikipedia.org/wiki/Python_(langage)')
print(article_data)
```
